```python
import pandas as pd

```

```python
df = pd.read_csv('../data/stockholm_monthly_mean_temperature_1756_2020_nonhom.csv')
df.head()
```

```python
df.columns
```

```python
df1 = df[['year', 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep',
       'oct', 'nov', 'dec']]

df1.head()
```

```python
df2 = pd.melt(df, id_vars='year', value_vars=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep',
       'oct', 'nov', 'dec'], var_name='month', value_name='temp')

df2
```

```python
month_list = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep',
       'oct', 'nov', 'dec']

month_dict = { month: i+1 for i, month in enumerate(month_list) }

df3 = df2.copy()

for month, num in month_dict.items():
    
    df3 = df3.replace(month, num)

df3
```

```python
df3['date'] = df3['year'].astype('str') + '-' + df3['month'].astype('str')
df3['date'] = pd.to_datetime(df3['date'])
df3[['date', 'temp']].to_csv('../data/stockholm-monthly-temp.csv',index=False)
```
